```python
hour_df.rename(columns={'instant':'rec_id',
                        'dteday':'datetime',
                        'holiday':'is_holiday',
                        'workingday':'is_workingday',
                       'weathersit':'weather_condition',
                       'hum':'humidity',
                       'mnth':'month',
                       'cnt':'total_count',
                       'yr':'year'},inplace=True)
```

```python
day_df.rename(columns={'instant':'rec_id',
                        'dteday':'datetime',
                        'holiday':'is_holiday',
                        'workingday':'is_workingday',
                       'weathersit':'weather_condition',
                       'hum':'humidity',
                       'mnth':'month',
                       'cnt':'total_count',
                       'yr':'year'},inplace=True)
```

```python
days = day_df['datetime'].str[-2:]
day_df["day"] = days
```

Data sudah bersih, lanjut ke Exploratory Data Analysis

## Exploratory Data Analysis

**Bike sharing per year**

```python
day_df.groupby(by="year").agg({
    "total_count":["max", "min", "mean", "std", "sum"]
})
```

**Bike Sharing in Working day, Holiday, and Weekday**

```python
day_df.groupby("year").agg({
    "is_holiday":"sum",
    "is_workingday":"sum",
    "weekday":"sum"
})
```

**bike sharing (2011 and 2012)**

```python
df_2011 = day_df[day_df["year"] ==0]
```

```python
df_2011
```

```python
df_2012 = day_df[day_df["year"] == 1]
df_2012.reset_index()
```

```python
df_2012
```

**see how many people use bike sharing every month in 2011**

```python
df_2011.groupby("month").agg({"total_count":"sum"})
```

**people using bike sharing in weekday, holiday, and working day per month in 2011**

```python
df_2011.groupby(by="month").agg({
    "weekday":"sum",
    "is_holiday":"sum",
    "is_workingday":"sum"
})
```

**See how many people using bike sharing in 2012 per month**

```python
df_2012.groupby("month").agg({"total_count":"sum"})
```

**people using bike sharing in weekday, holiday, and working day per month in 2012**

```python
df_2012.groupby(by="month").agg({
    "weekday":"sum",
    "is_holiday":"sum",
    "is_workingday":"sum"
})
```

# Data Visualization

```python
plt.figure(figsize=(10,5))

df_2011.groupby("month").agg({"total_count":"sum"}).plot(marker='o', linewidth=2,color="#72bcd4")

plt.title("Total Bike Sharing per Month (2011)", loc="center", fontsize=20)
plt.xlabel("month")
plt.ylabel("total count")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
df_2012.groupby("month").agg({"total_count":"sum"}).plot(kind='barh', color="#72bcd4")
plt.title('Total Bike Sharing per Month (2012)', fontsize=20)
plt.show()
```

```python
colors=["#72BCD4","#D3D3D3"]
sns.countplot(x='is_workingday', data=df_2011, palette=colors)
plt.title('Bike Sharing in Working Day (2011)', fontsize=18)
plt.xlabel("Working Day")
plt.ylabel("Total Count")
plt.show()
```

```python
colors=["#D3D3D3", "#72BCD4"]
sns.countplot(x='is_workingday', data=df_2012, palette=colors)
plt.title('Bike Sharing in Working Day (2012)', fontsize=18)
plt.xlabel("Working Day")
plt.ylabel("Total Count")
plt.show()
```
